```python
from dputils.scrape import Scraper, Tag
```

```python
url='https://www.flipkart.com/apple-iphone-15-blue-128-gb/p/itmbf14ef54f645d?pid=MOBGTAGPAQNVFZZY&lid=LSTMOBGTAGPAQNVFZZYO7HQ2L&marketplace=FLIPKART&store=tyy/4io&spotlightTagId=BestsellerId_tyy/4io&srno=b_1_1&otracker=browse&fm=organic&iid=2f9cd270-c66b-41b4-9f6a-ac1e8662a95f.MOBGTAGPAQNVFZZY.SEARCH&ppt=browse&ppn=browse&ssid=8u3h9dm6000000001712383095550'
sc = Scraper(url)
```

```python
title = Tag('span', cls='B_NuCI')
price = Tag(cls='_30jeq3 _16Jk6d')
sc.get_page_data(
    title=title, 
    price=price,
    rating_count= Tag('span', cls='_2_R_DZ'),
    config = Tag(cls='_2418kt')             
)
```

data extraction : Spider

```python
url ='https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
sc = Scraper(url)
```

```python
result = sc.get_repeating_page_data(
    target=Tag('body'),
    items=Tag(cls='_2kHMtA'),
    title = Tag(cls='_4rR01T'),
    price = Tag(cls='_30jeq3 _1_WHN1'),
    oprice = Tag(cls='_3I9_wc _27UcVY'),
    rating_n_review=Tag('span', cls='_2_R_DZ'),
    link = Tag('a', cls='_1fQZEK', output='href')
)    
```

```python
import pandas as pd
pd.DataFrame(result)
```
